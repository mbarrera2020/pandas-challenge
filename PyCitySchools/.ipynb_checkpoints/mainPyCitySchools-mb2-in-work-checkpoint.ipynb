```python
# Calculate the total number of students -- validation count: 39,170
    #TEMP
    #num_students = ss_df["Student ID"].count 
    #ss_df["Student ID"].count 
    
num_students = len(pd.unique(ss_df["Student ID"]))

#format data
num_students_f = "{:,.0f}".format(num_students)
print("Total number of students: ", num_students_f)
```

```python
# Calculate the total budget
total_budget = ss_df["budget"].sum()

#format data
total_budget_f = "${:,.2f}".format(total_budget)
print("Total budget: ", total_budget_f)
```

```python
# Calculate the average math score
avg_math_score = ss_df["math_score"].mean()

#format data
avg_math_score_f = "{:,.2f}".format(avg_math_score)
print("Average Math Score: ", avg_math_score_f)
```

```python
# Calculate the average reading score
avg_reading_score = ss_df["reading_score"].mean()

#format data
avg_reading_score_f = "{:,.2f}".format(avg_reading_score)
print("Average Reading Score: ", avg_reading_score_f)
```

```python
# FOR REFERENCE ONLY -- to get easy access of column / field names
ss_df.columns
```

```python
# MATH -- Calculate the percentage of students with a passing math score (70 or greater)
# 3 steps:

# USING DATAFRAME -- Subset of students with math score >= 70

# (1) create new df with only students with math score >= 70
st_math_score_70up_df = ss_df.loc[ss_df["math_score"] >=70, :]
st_math_score_70up_df.head()
```

```python
# MATH -- Goal: Calculate the percentage of students with a passing math score (70 or greater)
# 3 steps:
# (1)  total # of students = 39,170
```

```python
# Calculate the total number of students with math score >= 70 -- validation count: 29,370

num_students_ms70up = len(pd.unique(st_math_score_70up_df["Student ID"]))

#format data
num_students_ms70up_f = "{:,.0f}".format(num_students_ms70up)
print("Total number of students with math score >= 70: ", num_students_ms70up_f)
```

```python
# (3) Calculate the percentage of students with a passing math score (70 or greater) -- validation count: 74.98

pcnt_students_math_score70up = (num_students_ms70up / num_students) * 100

#format data
pcnt_students_math_score70up_f = "{:,.2f}".format(pcnt_students_math_score70up)
print("% students with a passing math score (70 or greater): ", pcnt_students_math_score70up_f)
```

```python
# READING -- Calculate the percentage of students with a passing reading score (70 or greater)
# 3 steps:

# USING DATAFRAME -- Subset of students with reading score >= 70

# (1) create new df with only students with reading score >= 70
st_reading_score_70up_df = ss_df.loc[ss_df["reading_score"] >=70, :]
st_reading_score_70up_df.head()
```

```python
# Calculate the total number of students with reading score >= 70 -- validation count: 33,610

num_students_rs70up = len(pd.unique(st_reading_score_70up_df["Student ID"]))

#format data
num_students_rs70up_f = "{:,.0f}".format(num_students_rs70up)
print("Total number of students with reading score >= 70: ", num_students_rs70up_f)
```

```python
# (3) Calculate the percentage of students with a passing reading score (70 or greater) -- validation: 85.81%

pcnt_students_reading_score70up = (num_students_rs70up / num_students) * 100

#format data
pcnt_students_reading_score70up_f = "{:,.2f}".format(pcnt_students_reading_score70up)
print("% students with a passing reading score (70 or greater): ", pcnt_students_reading_score70up_f)
```

```python
# Merge two dataframes using a right join 
merge_all_mr_df = pd.merge(st_math_score_70up_df, st_reading_score_70up_df, on="Student ID", how="right")
merge_all_mr_df
```

```python
# Remove the rows with missing data
merge_mr_df = merge_all_mr_df.dropna(how="any")
merge_mr_df
```

```python
# Calculate the total number of students -- validation count: 25,528

num_students_merge_mr = len(pd.unique(merge_mr_df["Student ID"]))

#format data
num_students_merge_mr_f = "{:,.0f}".format(num_students_merge_mr)
print("Total number of students with math & reading scores >= 70: ", num_students_merge_mr_f)
```

```python
pcnt_students_mr_score70up = (num_students_merge_mr / num_students) * 100

#format data
pcnt_students_mr_score70up_f = "{:,.2f}".format(pcnt_students_mr_score70up)
print("% students with a passing math & reading scores (70 or greater): ", pcnt_students_mr_score70up_f)
```

```python
# Create a dataframe to hold the above results

district_summary=[num_schools, num_students_f, total_budget_f, avg_math_score_f]

#district_summary=[num_schools, num_students_f, total_budget_f, avg_math_score_f, avg_reading_score_f, 
                  #pcnt_students_math_score70up_f, pcnt_students_reading_score70up_f, pcnt_students_mr_score70up_f]
                  
district_summary                 
```
